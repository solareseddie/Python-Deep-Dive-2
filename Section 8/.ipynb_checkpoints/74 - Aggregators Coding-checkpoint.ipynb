```python
def squares(n):
    for i in range(n):
        yield i**2
```

```python
list(squares(5))
```

```python
min(squares(5))
```

```python
max(squares(5))
```

```python
sum(squares(5))
```

```python
sq = squares(5)
```

```python
min(sq)
```

```python
max(sq)
```

```python
next(sq)
```

```python
class MySeq:
    def __init__(self, n):
        self.n = n

    def __len__(self):
        return self.n

    def __getitem__(self):
        pass
```

```python
my_seq = MySeq(0)
```

```python
bool(my_seq)
```

```python
my_seq = MySeq(10)
```

```python
bool(my_seq)
```

```python
l = [10, 20, 30, 40, 'hello']

is_all_numbers = True
for item in l:
    if not isinstance(item, Number):
        is_all_numbers = False
        break
```

```python
is_all_numbers
```

```python
l = [10, 20, 30, 40, 0]

all(l)
```

```python
def is_numeric(v):
    return isinstance(v, Number)
```

```python
pred_l = map(is_numeric, l)
print(list(pred_l))
```

```python
pred_l = (is_numeric(item) for item in l)
print(list(pred_l))
```

```python
pred_l = map(lambda x: isinstance(x, Number), l)
print(list(pred_l))
```

```python
all(pred_l)
```

```python
l = [10, 20, 30, 0, 'hello']
all(map(lambda x: isinstance(x, Number), l))
```

```python
with open('car-brands.txt') as f:
    for row in f:
        print(len(row), row, end='')
```

```python
with open('car-brands.txt') as f:
    result = all(map(lambda row: len(row) >= 4, f))

print(result)
```

```python
with open('car-brands.txt') as f:
    result = any(map(lambda row: len(row) > 10, f))

print(result)
```

```python
with open('car-brands.txt') as f:
    result = any(map(lambda row: len(row) > 13, f))

print(result)
```

```python
with open('car-brands.txt') as f:
    result = all((len(row) >= 4 for row in f))

print(result)
```
