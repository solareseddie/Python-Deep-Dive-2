```python
class MyClass:
    def __init__(self, name, val):
        self.name = name
        self.val = val

    def __repr__(self):
        return f'MyClass({self.name}, {self.val})'

    def __lt__(self, other):
        print('calling __lt__')
        return self.val < other.val
```

```python
c1 = MyClass('c1', 20)
c2 = MyClass('c2', 10)
c3 = MyClass('c3', 20)
c4 = MyClass('c4', 10)
```

```python
c1 < c2
```

```python
sorted([c1, c4, c3, c2])
```

```python
l = [c4, c2, c3, c1]
```

```python
l
```

```python
sorted(l, key=lambda c: c.name)
```
