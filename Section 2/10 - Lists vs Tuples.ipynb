Storage Efficiency

```python
import sys
```

```python
t = tuple()
prev = sys.getsizeof(t)

for i in range(10):
    c = tuple(range(i+1))
    size_c = sys.getsizeof(c)
    delta, prev = size_c - prev, size_c
    print(f'{i+1} items: {size_c}, delta = {delta}')
```

```python
l = list()
prev = sys.getsizeof(l)

for i in range(10):
    c = list(range(i+1))
    size_c = sys.getsizeof(c)
    delta, prev = size_c - prev, size_c
    print(f'{i+1} items: {size_c}, delta = {delta}')
```

```python
c = list()
prev = sys.getsizeof(c)
print(f'0 items: {prev}')

for i in range(255):
    c.append(i)
    size_c = sys.getsizeof(c)
    delta, prev = size_c - prev, size_c
    print(f'{i+1} items: {size_c}, delta = {delta}')
```
