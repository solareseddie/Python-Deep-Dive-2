```python
from functools import lru_cache

@lru_cache(2**10)
def fib(n):
    if n < 2:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)
```

```python
fib(5)
```

```python
fib(100)
```

```python
class Fib:
    def __init__(self, n):
        self.n = n

    def __len__(self):
        return self.n

    def __getitem__(self, s):
        if isinstance(s, int):
            if s < 0 or s >= self.n:
                raise IndexError
            else:
                return Fib._fib(s)

    @staticmethod
    @lru_cache(2**10)
    def _fib(n):
        if n < 2:
            return 1
        else:
            return fib(n - 1) + fib(n - 2)
```

```python
class Fib:
    def __init__(self, n):
        self.n = n

    def __len__(self):
        return self.n

    def __getitem__(self, s):
        if isinstance(s, int):
            if s < 0:
                s = self.n + s
            if s < 0 or s >= self.n:
                raise IndexError
            else:
                return Fib._fib(s)
        else:
            start, stop, step = s.indices(self.n)
            rng = range(start, stop, step)
            # rng = range(*s.indices(self.n))
            return [Fib._fib(i) for i in rng]
            
    @staticmethod
    @lru_cache(2**10)
    def _fib(n):
        if n < 2:
            return 1
        else:
            return Fib._fib(n - 1) + Fib._fib(n - 2)
```
