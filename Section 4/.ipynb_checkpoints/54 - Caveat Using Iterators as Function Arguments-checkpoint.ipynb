Caveat: Using Iterators as Function Arguments

```python
import random
```

```python
class Randoms:
    def __init__(self, n):
        self.n = n
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.i >= self.n:
            raise StopIteration
        else:
            self.i += 1
            return random.randint(0, 100)
```

```python
random.seed(0)
l = list(Randoms(10))
print(l)
```

```python
min(l), max(l)
```

```python
random.seed(0)
l = Randoms(10)
```

```python
min(l), max(l)
```

```python
f = open('cars.csv')
for row in f:
    print(row, end='')
f.close()
```

```python
def parse_data_row(row):
    row = row.strip('\n').split(',')
    return row[0], float(row[1])
```

```python
def max_mpg(data):
    max_mpg = 0
    for row in data:
        _, mpg = parse_data_row(row)
        if mpg > max_mpg:
            max_mpg = mpg
    return max_mpg
```

```python
f = open('cars.csv')
next(f)
next(f)
print(max_mpg(f))
f.close()
```

```python
def list_data(data, mpg_max):
    for row in data:
        car, mpg = parse_data_row(row)
        mpg_perc = mpg / mpg_max * 100
        print(f'{car}: {mpg_perc:.2f}%')
```

```python
f = open('cars.csv')
next(f), next(f)
list_data(f, 46.6)
f.close()
```

```python
with open('cars.csv') as f:
    next(f)
    next(f)
    list_data(f, 46.6)
```

```python
with open('cars.csv') as f:
    next(f)
    next(f)
    max_ = max_mpg(f)
    print(max_)
    list_data(f, max_)
```

```python
def list_data(data):
    max_mpg = 0
    for row in data:
        _, mpg = parse_data_row(row)
        if mpg > max_mpg:
            max_mpg = mpg

    for row in data:
        car, mpg = parse_data_row(row)
        mpg_perc = mpg / max_mpg * 100
        print(f'{car}: {mpg_perc:.2f}%')
```

```python
with open('cars.csv') as f:
    next(f), next(f)
    list_data(f)
```

```python
def list_data(data):
    if iter(data) is data:
        raise ValueError('data cannot be an iterator')
    max_mpg = 0
    for row in data:
        _, mpg = parse_data_row(row)
        if mpg > max_mpg:
            max_mpg = mpg

    for row in data:
        car, mpg = parse_data_row(row)
        mpg_perc = mpg / max_mpg * 100
        print(f'{car}: {mpg_perc:.2f}%')
```

```python
with open('cars.csv') as f:
    next(f), next(f)
    list_data(f)
```

```python
def list_data(data):
    if iter(data) is data:
        data = list(data)
        
    max_mpg = 0
    for row in data:
        _, mpg = parse_data_row(row)
        if mpg > max_mpg:
            max_mpg = mpg

    for row in data:
        car, mpg = parse_data_row(row)
        mpg_perc = mpg / max_mpg * 100
        print(f'{car}: {mpg_perc:.2f}%')
```

```python
with open('cars.csv') as f:
    next(f), next(f)
    list_data(f)
```
