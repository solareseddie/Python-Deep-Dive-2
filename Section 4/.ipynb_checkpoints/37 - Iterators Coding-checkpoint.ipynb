```python
class Squares:
    def __init__(self, length):
        self.i = 0
        self.length = length

    def __next__(self):
        if self.i >= self.length:
            raise StopIteration
        else:
            result = self.i ** 2
            self.i += 1
            return result

    def __iter__(self):
        return self
```

```python
class Squares:
    def __init__(self, length):
        self.i = 0
        self.length = length

    def __next__(self):
        print('__next__ called')
        if self.i >= self.length:
            raise StopIteration
        else:
            result = self.i ** 2
            self.i += 1
            return result

    def __iter__(self):
        print('__iter__ called')
        return self
```

```python
sq = Squares(5)

while True:
    try:
        item = next(sq)
        print(item)
    except StopIteration:
        break
```

```python
sq = Squares(5)
for item in sq:
    print(item)
```

```python
sq = Squares(5)
```

```python
[item for item in sq if item % 2 == 0]
```

```python
sq = Squares(5)
sq_iterator = iter(sq)
print(id(sq), id(sq_iterator))

while True:
    try:
        item = next(sq_iterator)
        print(item)
    except StopIteration:
        break
```
