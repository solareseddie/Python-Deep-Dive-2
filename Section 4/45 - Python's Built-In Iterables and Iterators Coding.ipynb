```python
f = open('cars.csv')
print(next(f))
print(f.__next__())
print(f.readline())

f.close()
```

```python
with open('cars.csv') as f:
    for row in f:
        print(row, end='')
```

```python
with open('cars.csv') as f:
    print(type(f))
    print('__iter__' in dir(f))
    print('__next__' in dir(f))
```

```python
with open('cars.csv') as f:
    print(iter(f) is f)
```

```python
with open('cars.csv') as f:
    l = f.readlines()
```

```python
l
```

```python
origins = set()

with open('cars.csv') as f:
    rows = f.readlines()

for row in rows[2:]:
    origin = row.strip('\n').split(',')[-1]
    origins.add(origin)

print(origins)
```

```python
origins = set()

with open('cars.csv') as f:
    next(f)
    next(f)
    for row in f:
        origin = row.strip('\n').split(',')[-1]
        origins.add(origin)

print(origins)
```
